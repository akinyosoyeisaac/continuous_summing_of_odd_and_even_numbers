```
s_odd = 0
s_even = 0
 
while True:
    try:
        numb = int(input('enter a number '))
        if numb == 0:
            print(f'The number {numb}')
        elif (numb % 2 == 0 and numb != 0):
            s_even += numb
            print(f'The number is an even number and the total sum is {s_even}')
        elif  (numb % 2 != 0 and numb != 0):
            s_odd += numb   
            if s_odd % 2 == 0:     
                print(f'The number is an even number and the total sum is {s_odd}')
            else:
                print(f'The number is an odd number and the total sum is {s_odd}')
    except:
        print('Check your input and start again')
    break
```
